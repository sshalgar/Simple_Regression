```python
# Converting Output to int values (Refining the output)
```

```python
iris_mapping = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2 }
iris["species"]= iris["species"].map(iris_mapping)
iris.tail()
```

```python
# Defining Inputs and Output (data)
```

```python
X = iris[["petal_length","petal_width","sepal_length","sepal_width"]].values
Y = iris["species"].values
```

```python
# Splitting the data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
len(x_train),len(x_test)
```

```python
# Loading the data into the Logistic Regression model
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X,Y)
```

```python
# Checking the accuracy of the model
```

```python
model.score(X,Y)
```

```python
# Making Preditions from Inputs
```

```python
expected = Y
predicted = model.predict(X)
predicted
```

```python
# Printing the report of the model's accuracy
```

```python
print (metrics.classification_report(expected, predicted))
```

```python
print (metrics.confusion_matrix(expected, predicted))
```

```python
# Visualizing the predicted and actual/expected values
```

```python
import matplotlib.pyplot as plt
plt.scatter(expected,predicted,c='blue')
plt.xlabel('Expected')
plt.ylabel('Predicted')
plt.grid()
```
